## Fill Null Values

```python
df.fillna(0, inplace=True)
df["Humidity"].mean()
```

```python
df.fillna(0, inplace=True)
df['Humidity'].mode()
```

```python
new_df = df.fillna(1)
new_df.head()
```

```python
# For replacing null values for different columns

new_df = df.fillna({
    'WindSpeedMPH':0,
    'Events':'No event'
})
new_df.head()
```

```python
# Replacing null values with previous value- forward fill

new_df = df.fillna(method="ffill")
new_df.head()
```

```python
# Replacing null values with previous value- backward fill

new_df = df.fillna(method="bfill")
new_df.head()
```

```python
new_df = df.dropna()
new_df
```

# Working with Skikit Datasets

```python
# From sklearn we are importing a dataset 

from sklearn.datasets import fetch_california_housing
boston_dataset = fetch_california_housing()
type(boston_dataset)
```

```python
print(boston_dataset )
```

```python
# Dataset that we loaded is putted on data frame

boston_df =pd.DataFrame(boston_dataset.data, columns=boston_dataset.feature_names)
boston_df.head(8)
```

```python
# To download csv file that we already modified
# To download excel file

boston_df.to_csv('boston.csv')
# df.to_excel("df")
```

```python
# To find the missing values

boston_df.isnull()
```

```python
boston_df.value_counts('HouseAge')
```

```python
boston_df.isnull().sum()
```
